```python
#pseudocode
L = [10, 1, 2, -3, 5, -2]

def searchMinFromList(L,n):
    minValue = L[0] 
    counter = 1
    while (counter<n):
        v = L[counter]
        if (v < minValue):
            minValue = v
            idx = counter
        else:
            pass

return minValue, idx
    
```

```python
#sortListPseudocode

def sortList(L,n):
    L2 =[]
    counter = 0
    while(counter<=n):
        min, idx = searchMinFromList(L,n)
        L2.append(min)
        del L[idx]
        n = n-1
    
return L2
L2 = [-3, -2, 1, 2,5, 10]

    
```

```python
"""
Given a list [1,2,4,-5,7,9,3,2], make another list
that contains all the items in increasing sorted order ie, [-5,1,2,2,3,4,7,9]
"""
L = [1,2,4,-5,7,9,3,2]
for j in range(len(L)):
    m = L[j]
    idx = j
    c = j
    for i in range(j, len(L)):
        if L[i]<m:
            m = L[i]
            idx = c
        c += 1
    tmp = L[j]
    L[j] = m
    L[idx] = tmp
L

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

def COSDensity(cf,x,N,a,b):
    i = complex(0.0,1.0) 
    k = np.linspace(0,N-1,N)
    u = np.zeros([1,N])
    u = k * np.pi / (b-a)
        
    F_k    = 2.0 / (b - a) * np.real(cf(u) * np.exp(-i * u * a));
    F_k[0] = F_k[0] * 0.5; 
    
    f_X = np.matmul(F_k , np.cos(np.outer(u, x - a )))

    return f_X
    
def mainCalculation():
    i = complex(0.0, 1.0) 
    a = -10
    b = 10
    N = [16, 64, 128]

    mu = 0.5
    sigma = 0.2 
        
    cF = lambda u : np.exp(i * mu * u - 0.5 * np.power(sigma,2.0) * np.power(u,2.0));
        
    y = np.linspace(0.05,5,1000)
        
    plt.figure(1)
    plt.grid()
    plt.xlabel("y")
    plt.ylabel("$f_Y(y)$")
    for n in N:
        f_Y = 1/y * COSDensity(cF,np.log(y),n,a,b)

        plt.plot(y,f_Y)
    plt.legend(["N=%.0f"%N[0],"N=%.0f"%N[1],"N=%.0f"%N[2]])
    
mainCalculation()
```

```python
def checkIfNotNumber2(L):
    for x in L:
        if not(isinstance(x,(float,int))):
            return False
    return True
checkIfNotNumber(12)
```

```python
def findMin(L,startIndx):
    min = L[startIndx] #variable to store minimum value
    idx = startIndx #counts index position
    for i in range(startIndx,len(L)):
        x = L[i]
        if x<min:
            min=x
            idx=i
        else:
            pass
    return min,idx #returns the value we are looking for 
#a,b=findMin([20,19,18,1,0])
#print(a,b)
```

```python
def swapValues(L,idx1,idx2):
    tmp=L[idx1]
    L[idx1]=L[idx2]
    L[idx2]=tmp
    return L
L=[1,2,3,4,5]
L2 = swapValues(L,2,4)
print(L2)
```

```python
def sortList(L):
    if not(checkIfNotNumber2(L)):
        print("Error: List does not contain numeric values")
        return
    else:
        c = 0
        for x in L:
            m,idx=findMin(L,c)
            L = swapValues(L,c,idx)
            c+=1
    return L
        
```

```python
"""
Statement Problem: lets say you are a teach and you have different student records containing ID
of a student and the grades list in each subject where different students have
taken different number of subjects. All these records are in hard copy. You want to enter all 
the data in computer and want to compute the average grades of each student and display

"""
def getDataFromUsers():
    D = {}
    while True:
        studentId = input("Enter student ID: ")
        marksList = input("Enter the marks by comma separated values: ")
        moreStudents = input("Enter yes/no for adding more students: ")
        if studentId in D: 
            print(studentId,"is already inserted")
        else:
            D[studentId] = marksList.split(",")
        if moreStudents.lower() == "no":
            return D
            
```

```python
studentData = getDataFromUsers()
```

```python
studentData
```

```python
#average marks of each students
def getAvgMarks(D):
    avgMarks = {}
    for x in D:
        L = D[x]
        s = 0
        for marks in L:
            s += int(marks)
        avgMarks[x] = s/len(L)
    return avgMarks
```

```python
#shows the trend of confirmed, recovered, and deaths for all 183 countries
for idx in range(0,len(countries)): #loop over each country
    C = df3[df3['Country']==countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)),C['Confirmed'],color='blue',label='Confirmed')
    plt.scatter(np.arange(0,len(C)),C['Recovered'],color='green',label='Recovered')
    plt.scatter(np.arange(0,len(C)),C['Deaths'],color='red',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the first case discovered')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.show()
    
```
